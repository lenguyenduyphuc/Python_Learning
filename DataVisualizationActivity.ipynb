```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Option 1: Fewer bins (8 compass directions)
bins_many = [0, 22.5, 45, 67.5, 90, 112.5, 135, 157.5, 180,
             202.5, 225, 247.5, 270, 292.5, 315, 337.5, 360]

df['Wind Direction Category'] = pd.cut(df['Wind Direction'], bins=bins_many)
# Plot with fewer bins (8 directions)
plt.figure(figsize=(12, 8))
sns.barplot(x='Wind Direction Category', y='Wind Speed', data=df)
plt.ylabel('Average Wind Speed (m/s)')
plt.xlabel('Wind Direction')
plt.title('Average Wind Speed by Wind Direction (16 Directions)', color='blue', fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
