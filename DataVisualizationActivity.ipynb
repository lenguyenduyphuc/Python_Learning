```python
# Import the csv file
# upload fatal-police-shootings-data.csv file
from google.colab import files
uploaded = files.upload()
```

```python
# convert csv file to pandas format
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['Marine_CSV_sample (3).csv']))
print(df)
```

```python
# Display the columns for this dataset
print(df.columns)
```

```python
# Drop columns where all values are NaN
df = df.dropna(axis=1, how='all')
```

```python
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Count plot of Ice Accretion On Ship
drop_data = {}
plt.figure(figsize=(8, 8))
sns.countplot(x='Cloud Height', data=df)
plt.title('Cloud Height')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Option 1: Fewer bins (8 compass directions)
bins_many = [0, 22.5, 45, 67.5, 90, 112.5, 135, 157.5, 180,
             202.5, 225, 247.5, 270, 292.5, 315, 337.5, 360]

df['Wind Direction Category'] = pd.cut(df['Wind Direction'], bins=bins_many)
# Plot with fewer bins (8 directions)
plt.figure(figsize=(12, 8))
sns.barplot(x='Wind Direction Category', y='Wind Speed', data=df)
plt.ylabel('Average Wind Speed (m/s)')
plt.xlabel('Wind Direction')
plt.title('Average Wind Speed by Wind Direction (16 Directions)', color='blue', fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print(df.columns.tolist())
```

```python
sns.histplot(x="Wind Speed", data=df, bins=15)
plt.ylabel('Frequency')  # Changed from 'Wind Speed' to 'Frequency' since y-axis shows counts
plt.xlabel('Wind Speed')  # Added x-label for clarity
plt.title('Wind Speed Distribution', color='blue', fontsize=15)  # Changed title to match the data
plt.xticks(rotation=45)
plt.tight_layout()  # Added for better layout
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="Air Temperature", fill=True)
plt.title('Density of Air Temperature', fontsize=15)
plt.xlabel('Temperature')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Wind Direction Category", y="Wind Speed")
plt.title('Wind Speed by Direction Category', fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Rate of Ice Accretion on Ship", y="Sea Level Pressure")
plt.title('Rate of Ice Accretion vs Sea Level Pressure', fontsize=15)
plt.xlabel('Rate of Ice Accretion on Ship')
plt.ylabel('Sea Level Pressure')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Sea Level Pressure", y="Wind Speed")
plt.title('Sea Level Pressure vs Wind Speed', fontsize=15)
plt.xlabel('Sea Level Pressure')
plt.ylabel('Wind Speed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
